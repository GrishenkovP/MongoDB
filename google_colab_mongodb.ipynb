```
# Подключаем библиотеки
import pymongo
from pymongo import MongoClient
import datetime
from pprint import pprint
```

```
# Соединение с сервером базы данных
client = MongoClient()
```

```
# Получение доступа к базе данных
db = client.db_test
```

```
project = {"name": "Основная БД",
           "description": "Разворачивание основной БД",
           "tags": ["db", "postgesql", "sql"],
           "date": datetime.datetime.utcnow()}
```

```
# Добавление одного документа в коллекцию
projects = db.projects
projects.insert_one(project)
```

```
# Проверяем корректно ли записался документ
pprint(db.projects.find()[0])
```

```
# Массовое добавление документов в коллекцию
project_list = [{"name": "Вспомогательная БД",
                 "description": "Разворачивание ClickHouse для хранения логов пользователей игры",
                 "tags": ["db", "clickhouse", "nosql"],
                 "date": datetime.datetime.utcnow()},
                {"name": "Сервер",
                "description": "Аренда сервера с Debian",
                "tags": ["server", "linux","debian"],
                "date": datetime.datetime.utcnow()},
                {"name": "ETL",
                "description": "Написание основных дагов по платформе Apache Airflow",
                "tags": ["python", "etl", "apache_airflow"],
                "date": datetime.datetime.utcnow()},
            {"name": "BI аналитика",
             "description": "Построение комплексных дашбордов с метриками для оценки действий пользователей портала",
             "tags": ["bi", "data_science", "tableau"],
             "date": datetime.datetime.utcnow()},
           ]
projects.insert_many(project_list)
```

```
# Получение первого документа из коллекции
project_first = projects.find_one()
pprint(project_first)
```

```
# Перебрать все документы
for _ in projects.find():
    pprint(_)
```

```
# Поиск документа по условию
pprint(db.projects.find({"name": "Сервер"})[0])
```

```
# Сортировка записей по условию
for _ in projects.find({"tags":"db"}).sort("name"):
    pprint(_)
```

```
# Выбор столбцов для вывода (1/0)
for _ in projects.find({},{"name":1,"_id":0}).sort("name"):
    pprint(_)
```

```
# Вывести только первые три документа
for _ in db.projects.find().limit(3):
  pprint(_)
```

```
# Подсчет количества документов в коллекции
print(db.projects.count())
```

```
# Найти только уникальные значения для поля
list_tags =  db.projects.distinct("tags")
print(list_tags)
```

```
# Условные операторы
# $eq (равно)
# $ne (не равно) 
# $gt (больше чем)
# $lt (меньше чем)
# $gte (больше или равно)
# $lte (меньше или равно)
# $in определяет массив значений, одно из которых должно иметь поле документа
# $nin определяет массив значений, которые не должно иметь поле документа

# Логические операторы
# $or: соединяет два условия, и документ должен соответствовать одному из этих условий
# $and: соединяет два условия, и документ должен соответствовать обоим условиям
# $not: документ должен НЕ соответствовать условию
# $nor: соединяет два условия, и документ должен НЕ соответстовать обоим условиям

# $all определяет массив возможных выражений и требует, чтобы документы имели весь определяемый набор выражений
# $size используется для нахождения документов, в которых массивы имеют число элементов, равным значению $size
# $exists позволяет извлечь только те документы, в которых определенный ключ присутствует или отсутствует
# $type извлекает только те документы, в которых определенный ключ имеет значение определенного типа, например, строку или число
```

```
for _ in db.projects.find({"$or":[{"name": "Основная БД"}, {"tags": "bi"}]}):
  pprint(_)
```
